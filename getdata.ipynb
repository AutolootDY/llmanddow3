```python
import pandas as pd

# ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå CSV (‡∏™‡∏°‡∏°‡∏∏‡∏ï‡∏¥‡∏ß‡πà‡∏≤‡πÑ‡∏ü‡∏•‡πå‡∏ä‡∏∑‡πà‡∏≠ 'trend_data.csv')
df = pd.read_csv('llm_googletrand.csv')

# ‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'Date' ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏ô‡∏¥‡∏î datetime
df['Date'] = pd.to_datetime(df['Date'])

# ‡∏ï‡∏±‡πâ‡∏á 'Date' ‡πÄ‡∏õ‡πá‡∏ô index ‡∏Ç‡∏≠‡∏á DataFrame
df.set_index('Date', inplace=True)

# ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå
print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå:")
print(df)

# Resample ‡πÄ‡∏õ‡πá‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏ß‡∏±‡∏ô‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ Forward Fill
df_daily_ffill = df.resample('D').ffill()
print("\n‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏ß‡∏±‡∏ô (Forward Fill):")
print(df_daily_ffill)

# Resample ‡πÄ‡∏õ‡πá‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏ß‡∏±‡∏ô‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ Linear Interpolation
df_daily_interpolate = df.resample('D').interpolate(method='linear')
print("\n‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏ß‡∏±‡∏ô (Linear Interpolation):")
print(df_daily_interpolate)

df_daily_interpolate.to_csv("df_daily_interpolate.csv")
```

```python
import os
import glob
import pandas as pd

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏∏‡πâ‡∏ô‡πÅ‡∏•‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Google Trend
stock_folder = 'ndx_date'   # ‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏∏‡πâ‡∏ô Dow Jones 30 stocks
trend_folder = 'llmdate_copy'   # ‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Google Trend
output_folder = 'output_folder' # ‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå

os.makedirs(output_folder, exist_ok=True)

# ----- 1. ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå CSV ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏∏‡πâ‡∏ô -----
stock_files = glob.glob(os.path.join(stock_folder, '*.csv'))
stock_dfs = []
for file in stock_files:
    # ‡∏≠‡πà‡∏≤‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏∏‡πâ‡∏ô ‡πÇ‡∏î‡∏¢ parse ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'datetime'
    df = pd.read_csv(file, parse_dates=['datetime'])
    stock_dfs.append(df)
# ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
df_stock = pd.concat(stock_dfs, ignore_index=True)

# ----- 2. ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå CSV ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Google Trend -----
trend_files = glob.glob(os.path.join(trend_folder, '*.csv'))
trend_dfs = []
for file in trend_files:
    # ‡∏≠‡πà‡∏≤‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Google Trend ‡πÇ‡∏î‡∏¢ parse ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only'
    # ‡∏õ‡∏£‡∏±‡∏ö dayfirst=True ‡∏´‡∏≤‡∏Å‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö dd/mm/yyyy (‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: "3/1/2023" ‡πÉ‡∏´‡πâ‡πÅ‡∏õ‡∏•‡πÄ‡∏õ‡πá‡∏ô 2023-01-03)
    df = pd.read_csv(file, parse_dates=['date_only'], dayfirst=True)
    trend_dfs.append(df)
# ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Google Trend ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
df_trend = pd.concat(trend_dfs, ignore_index=True)

# ----- 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only' ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏∏‡πâ‡∏ô (‡∏ï‡∏±‡∏î‡πÄ‡∏ß‡∏•‡∏≤‡∏à‡∏≤‡∏Å 'datetime') -----
df_stock['date_only'] = df_stock['datetime'].dt.date

# ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö Google Trend ‡πÄ‡∏£‡∏≤‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô date (‡∏ñ‡πâ‡∏≤‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡∏ó‡∏≥)
df_trend['date_only'] = df_trend['date_only'].dt.date

# ----- 4. Merge ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏∏‡πâ‡∏ô‡∏Å‡∏±‡∏ö Google Trend ‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only' -----
# ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only' ‡πÅ‡∏•‡∏∞ 'llm' ‡∏à‡∏≤‡∏Å df_trend
df_merged = pd.merge(df_stock, df_trend[['date_only', 'llm']], on='date_only', how='inner')

# ‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only' ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô datetime (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£ filter ‡πÅ‡∏•‡∏∞ rolling)
df_merged['date_only'] = pd.to_datetime(df_merged['date_only'])
# ‡∏ï‡∏±‡πâ‡∏á index ‡πÄ‡∏õ‡πá‡∏ô 'date_only'
df_merged.set_index('date_only', inplace=True)

# ----- 5. ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏´‡πâ‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏ä‡πà‡∏ß‡∏á‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ -----
start_date = '2023-01-01'
end_date   = '2025-03-14'
df_merged = df_merged.loc[start_date:end_date]

# ----- 6. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Rolling Correlation (6 ‡πÄ‡∏î‡∏∑‡∏≠‡∏ô ~ 180 ‡∏ß‡∏±‡∏ô)180D ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô -----
def compute_rolling_corr(group, window='180D'):
    group = group.sort_index()  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏ï‡∏≤‡∏°‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì rolling correlation ‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 'close' ‡∏Å‡∏±‡∏ö 'llm'
    group['rolling_corr'] = group['close'].rolling(window=window).corr(group['llm'])
    return group

df_result = df_merged.groupby('symbol', group_keys=False).apply(compute_rolling_corr)

# ‡πÅ‡∏™‡∏î‡∏á‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
print("‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà merge ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì rolling correlation:")
print(df_result.head())

# ----- (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å) ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏•‡∏á‡πÑ‡∏ü‡∏•‡πå CSV -----
output_filepath = os.path.join(output_folder, 'merged_rolling_corr_adx.csv')
df_result.to_csv(output_filepath)
print(f"\nSaved merged and rolling correlation data to {output_filepath}")
```

```python
import plotly.graph_objects as go
import pandas as pd

def surface_3d_rolling_corr(df):
    """
    ‡∏™‡∏£‡πâ‡∏≤‡∏á 3D surface plot ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö rolling correlation
    ‡πÇ‡∏î‡∏¢ df ‡∏Ñ‡∏ß‡∏£‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only', 'symbol' ‡πÅ‡∏•‡∏∞ 'rolling_corr'
    """
    # Pivot DataFrame ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÑ‡∏î‡πâ grid ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö plot:
    # index (Y-axis): date_only, columns (X-axis): symbol, values (Z-axis): rolling_corr
    pivot_df = df.reset_index().pivot(index='date_only', columns='symbol', values='rolling_corr')
    
    # ‡∏™‡∏£‡πâ‡∏≤‡∏á figure 3D surface plot
    fig = go.Figure(data=[go.Surface(
        x=pivot_df.columns,       # ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô
        y=pivot_df.index,         # ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
        z=pivot_df.values,        # ‡∏Ñ‡πà‡∏≤ rolling correlation
        opacity=0.95,
        connectgaps=True,
        colorscale='rdbu',
        showscale=True,
        reversescale=True,
    )])
    
    # ‡∏õ‡∏£‡∏±‡∏ö layout ‡∏Ç‡∏≠‡∏á‡∏Å‡∏£‡∏≤‡∏ü
    fig.update_layout(
        title='3D Surface Plot of Rolling Correlation',
        title_font=dict(size=20),
        autosize=True,
        height=500,
        hovermode='closest',
        scene={
            "aspectratio": {"x": 1, "y": 2.2, "z": 1},
            'camera': {'eye': {'x': 2, 'y': 0.4, 'z': 0.8}},
            'xaxis_title': 'Symbol',
            'yaxis_title': 'Date',
            'zaxis_title': 'Rolling Correlation'
        },
        margin=dict(t=40),
        annotations=[
            dict(
                text="Data Source: Your Data Source",
                x=0,
                y=-0.15,
                xref="paper",
                yref="paper",
                showarrow=False
            )
        ]
    )
    return fig

# ‡∏™‡∏°‡∏°‡∏∏‡∏ï‡∏¥‡∏ß‡πà‡∏≤ df_result ‡∏Ñ‡∏∑‡∏≠ DataFrame ‡∏ó‡∏µ‡πà‡∏°‡∏µ rolling_corr ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only', 'symbol'
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô:
fig = surface_3d_rolling_corr(df_result)
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

def create_3d_surface_with_toggle(pivot_df):
    """
    ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü 3D ‡∏ó‡∏µ‡πà‡∏°‡∏µ update menu ‡πÉ‡∏´‡πâ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏î‡∏π rolling correlation ‡∏Ç‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô
    pivot_df ‡∏Ñ‡∏ß‡∏£‡πÄ‡∏õ‡πá‡∏ô DataFrame ‡∏ó‡∏µ‡πà pivot ‡πÇ‡∏î‡∏¢‡∏°‡∏µ index ‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
    ‡πÅ‡∏•‡∏∞ columns ‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô (symbol) ‡πÇ‡∏î‡∏¢‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô pivot ‡∏Ñ‡∏∑‡∏≠ rolling_corr
    """
    # ‡∏î‡∏∂‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô‡∏à‡∏≤‡∏Å pivot table
    symbols = list(pivot_df.columns)
    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÅ‡∏Å‡∏ô X ‡πÅ‡∏ö‡∏ö‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô
    x_numeric = np.arange(len(symbols))
    
    # --- Trace ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• "All" ---
    all_trace = go.Surface(
        x = x_numeric,
        y = pivot_df.index,
        z = pivot_df.values,
        opacity=0.95,
        colorscale='rdbu',
        reversescale=True,
        showscale=True,
        name="All"
    )
    
    traces = [all_trace]
    
    # --- Trace ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô (Scatter3d ‡πÅ‡∏ö‡∏ö‡πÄ‡∏™‡πâ‡∏ô) ---
    for i, symbol in enumerate(symbols):
        scatter_trace = go.Scatter3d(
            x = [x_numeric[i]] * len(pivot_df.index),
            y = pivot_df.index,
            z = pivot_df[symbol].values,
            mode='lines',
            name=symbol
        )
        traces.append(scatter_trace)
    
    # ‡∏™‡∏£‡πâ‡∏≤‡∏á Figure ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ö trace ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
    fig = go.Figure(data=traces)
    
    # --- ‡∏™‡∏£‡πâ‡∏≤‡∏á update menu buttons ---
    buttons = []
    # ‡∏õ‡∏∏‡πà‡∏° "All" ‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞ trace ‡πÅ‡∏ö‡∏ö Surface (trace ‡∏ó‡∏µ‡πà 0)
    visibility_all = [True] + [False] * len(symbols)
    buttons.append(dict(
        label="All",
        method="update",
        args=[{"visible": visibility_all},
              {"title": "Rolling Correlation - All Stocks"}]
    ))
    
    # ‡∏õ‡∏∏‡πà‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏∏‡πâ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß (trace ‡∏ó‡∏µ‡πà 1 ‡∏ñ‡∏∂‡∏á n)
    for i, symbol in enumerate(symbols):
        # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏õ‡∏∏‡πà‡∏° ‡πÉ‡∏´‡πâ‡∏ã‡πà‡∏≠‡∏ô trace ‡πÅ‡∏ö‡∏ö Surface (index 0) ‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞ trace ‡∏Ç‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡∏ô‡∏±‡πâ‡∏ô
        visibility = [False]  # ‡∏ã‡πà‡∏≠‡∏ô trace "All"
        # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö scatter traces: ‡πÉ‡∏´‡πâ trace ‡∏ó‡∏µ‡πà i+1 (‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏à‡∏≤‡∏Å index 1) ‡πÄ‡∏õ‡πá‡∏ô True
        for j in range(len(symbols)):
            visibility.append(j == i)
        buttons.append(dict(
            label=symbol,
            method="update",
            args=[{"visible": visibility},
                  {"title": f"Rolling Correlation - {symbol}"}]
        ))
    
    fig.update_layout(
        updatemenus=[{
            "buttons": buttons,
            "direction": "down",
            "showactive": True,
            "x": 0.1,
            "xanchor": "left",
            "y": 1.1,
            "yanchor": "top"
        }],
        title="Rolling Correlation 3D Plot",
        scene=dict(
            xaxis=dict(
                title="Stock Symbol",
                tickmode="array",
                tickvals=x_numeric,
                ticktext=symbols
            ),
            yaxis_title="Date",
            zaxis_title="Rolling Correlation"
        )
    )
    
    return fig

# ‡∏™‡∏°‡∏°‡∏∏‡∏ï‡∏¥‡∏ß‡πà‡∏≤ pivot_df ‡∏Ñ‡∏∑‡∏≠ DataFrame ‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£ pivot ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
# ‡πÄ‡∏ä‡πà‡∏ô ‡∏à‡∏≤‡∏Å DataFrame df_result ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'date_only', 'symbol' ‡πÅ‡∏•‡∏∞ 'rolling_corr'
# ‡πÄ‡∏£‡∏≤‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏™‡∏£‡πâ‡∏≤‡∏á pivot_df ‡πÑ‡∏î‡πâ‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á:
#
#   pivot_df = df_result.reset_index().pivot(index='date_only', columns='symbol', values='rolling_corr')
#
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡∏µ‡πâ‡πÉ‡∏ä‡πâ pivot_df ‡∏ó‡∏µ‡πà‡∏°‡∏µ index ‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà (DatetimeIndex) ‡πÅ‡∏•‡∏∞ columns ‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô

# ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü 3D ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏õ‡∏∏‡πà‡∏°‡πÄ‡∏õ‡∏¥‡∏î/‡∏õ‡∏¥‡∏î‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏´‡∏∏‡πâ‡∏ô
fig = create_3d_surface_with_toggle(df_result)
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

def plot_3d_lines_with_toggle(df_result):
    """
    ‡∏™‡∏£‡πâ‡∏≤‡∏á 3D line plot ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö rolling correlation ‡πÇ‡∏î‡∏¢‡∏°‡∏µ‡∏õ‡∏∏‡πà‡∏°‡∏™‡∏•‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô
    df_result ‡∏Ñ‡∏ß‡∏£‡∏°‡∏µ index ‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà ‡πÅ‡∏•‡∏∞‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'symbol' ‡πÅ‡∏•‡∏∞ 'rolling_corr'
    """
    # ‡∏î‡∏∂‡∏á‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥ ‡πÅ‡∏•‡πâ‡∏ß‡∏à‡∏±‡∏î‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏ô‡πà‡πÉ‡∏à
    symbols = sorted(df_result['symbol'].unique())
    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î mapping ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏Å‡∏ô y: ‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô‡∏à‡∏∞‡πÑ‡∏î‡πâ‡∏Ñ‡πà‡∏≤ y ‡πÄ‡∏â‡∏û‡∏≤‡∏∞ (0, 1, 2, ...)
    symbol_to_y = {symbol: i for i, symbol in enumerate(symbols)}
    
    # ‡∏™‡∏£‡πâ‡∏≤‡∏á trace ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô (Scatter3d ‡πÅ‡∏ö‡∏ö line)
    traces = []
    for symbol in symbols:
        # ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏∏‡πâ‡∏ô‡∏ô‡∏±‡πâ‡∏ô
        df_sym = df_result[df_result['symbol'] == symbol].sort_index()
        # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÅ‡∏Å‡∏ô x ‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà, ‡πÅ‡∏Å‡∏ô y ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏∏‡πâ‡∏ô‡∏ô‡∏±‡πâ‡∏ô, ‡πÅ‡∏Å‡∏ô z ‡πÄ‡∏õ‡πá‡∏ô rolling_corr
        x = df_sym.index  # index ‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
        y = [symbol_to_y[symbol]] * len(df_sym)
        z = df_sym['rolling_corr']
        
        trace = go.Scatter3d(
            x=x,
            y=y,
            z=z,
            mode='lines+markers',
            name=symbol
        )
        traces.append(trace)
    
    # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏õ‡∏∏‡πà‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏™‡∏•‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• (update menu)
    buttons = []
    
    # ‡∏õ‡∏∏‡πà‡∏° "All" ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏™‡∏î‡∏á‡∏ó‡∏∏‡∏Å‡∏´‡∏∏‡πâ‡∏ô
    all_visible = [True] * len(traces)
    buttons.append(dict(
        label="All",
        method="update",
        args=[{"visible": all_visible},
              {"title": "Rolling Correlation: All Stocks"}]
    ))
    
    # ‡∏õ‡∏∏‡πà‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô‡πÅ‡∏¢‡∏Å‡∏Å‡∏±‡∏ô
    for i, symbol in enumerate(symbols):
        visibility = [False] * len(traces)
        visibility[i] = True
        buttons.append(dict(
            label=symbol,
            method="update",
            args=[{"visible": visibility},
                  {"title": f"Rolling Correlation: {symbol}"}]
        ))
    
    layout = go.Layout(
        title="3D Rolling Correlation with Toggle Buttons",
        scene=dict(
            xaxis_title="Date",
            yaxis_title="Stock",
            zaxis_title="Rolling Correlation",
            # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î tick labels ‡πÉ‡∏´‡πâ‡πÅ‡∏Å‡∏ô y ‡∏ï‡∏≤‡∏° mapping
            yaxis=dict(
                tickmode='array',
                tickvals=list(symbol_to_y.values()),
                ticktext=list(symbol_to_y.keys())
            )
        ),
        updatemenus=[{
            "buttons": buttons,
            "direction": "down",
            "showactive": True,
            "x": 0.1,
            "xanchor": "left",
            "y": 1.1,
            "yanchor": "top"
        }]
    )
    
    fig = go.Figure(data=traces, layout=layout)
    return fig

# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
# ‡∏™‡∏°‡∏°‡∏∏‡∏ï‡∏¥‡∏ß‡πà‡∏≤ df_result ‡∏Ñ‡∏∑‡∏≠ DataFrame ‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ merge ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì rolling correlation ‡πÑ‡∏ß‡πâ‡πÅ‡∏•‡πâ‡∏ß
# ‡πÇ‡∏î‡∏¢ index ‡πÄ‡∏õ‡πá‡∏ô datetime (‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà) ‡πÅ‡∏•‡∏∞‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'symbol' ‡πÅ‡∏•‡∏∞ 'rolling_corr'

# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô
fig = plot_3d_lines_with_toggle(df_result)
fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

def surface_3d_with_toggle(df_result):
    """
    ‡∏™‡∏£‡πâ‡∏≤‡∏á 3D Surface Plot ‡∏Ç‡∏≠‡∏á Rolling Correlation
    ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏õ‡∏∏‡πà‡∏° (Toggle) ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏™‡∏î‡∏á 'All' ‡∏´‡∏£‡∏∑‡∏≠ 'Single Symbol'
    
    - df_result ‡∏Ñ‡∏ß‡∏£‡∏°‡∏µ:
        index: ‡πÄ‡∏õ‡πá‡∏ô date (date_only ‡∏´‡∏£‡∏∑‡∏≠ datetime)
        columns: 'symbol', 'rolling_corr'
    """
    # ‡∏£‡∏µ‡πÄ‡∏ã‡πá‡∏ï index ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå date ‡∏°‡∏µ‡∏ä‡∏∑‡πà‡∏≠‡∏ä‡∏±‡∏î‡πÄ‡∏à‡∏ô
    df_temp = df_result.reset_index().copy()
    # ‡∏ï‡∏±‡πâ‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå index ‡∏Ç‡∏≠‡∏á‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÉ‡∏´‡πâ‡∏ä‡∏±‡∏î‡πÄ‡∏à‡∏ô (‡∏ñ‡πâ‡∏≤‡πÉ‡∏ô df_result ‡πÄ‡∏õ‡πá‡∏ô date_only ‡∏≠‡∏¢‡∏π‡πà‡πÅ‡∏•‡πâ‡∏ß ‡πÉ‡∏´‡πâ‡∏õ‡∏£‡∏±‡∏ö‡∏ï‡∏≤‡∏°‡∏à‡∏£‡∏¥‡∏á)
    df_temp.rename(columns={'index': 'date_only'}, inplace=True)
    
    # Pivot ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥ Surface "All":
    #   index  = symbol
    #   columns= date
    #   values = rolling_corr
    pivot_all = df_temp.pivot(index='symbol', columns='date_only', values='rolling_corr')
    pivot_all = pivot_all.sort_index(axis=0)  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á symbol
    pivot_all = pivot_all.sort_index(axis=1)  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á date
    
    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏û‡∏≠‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
    if pivot_all.empty:
        raise ValueError("Pivot DataFrame ‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏õ‡∏•‡πà‡∏≤ ‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• df_result")
    
    # ‡∏™‡∏£‡πâ‡∏≤‡∏á trace ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö "All" (Surface ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏ó‡∏µ‡πà‡∏°‡∏µ Symbol ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î)
    trace_all = go.Surface(
        x = pivot_all.columns,        # X = Date
        y = pivot_all.index,          # Y = Symbol
        z = pivot_all.values,         # Z = Rolling Correlation
        colorscale='rdbu',
        reversescale=True,
        connectgaps=True,
        opacity=0.95,
        showscale=True,    # ‡πÅ‡∏™‡∏î‡∏á color bar ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö All
        name='All'
    )
    
    # ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏™‡∏£‡πâ‡∏≤‡∏á surface ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞ Symbol
    # ‡πÇ‡∏î‡∏¢‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö symbol ‡πÉ‡∏î‡πÜ ‡πÄ‡∏°‡∏∑‡πà‡∏≠ fix symbol ‡πÅ‡∏•‡πâ‡∏ß ‡∏à‡∏∞‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• date vs rolling_corr ‡πÄ‡∏õ‡πá‡∏ô 1 ‡πÅ‡∏ñ‡∏ß
    # ‡∏ï‡πâ‡∏≠‡∏á‡∏ó‡∏≥ pseudo-surface 2 ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
    single_traces = []
    symbols = pivot_all.index.tolist()  # ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô‡∏à‡∏≤‡∏Å pivot
    
    for symbol in symbols:
        # ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• rolling_corr ‡∏Ç‡∏≠‡∏á symbol ‡∏ô‡∏±‡πâ‡∏ô (‡∏à‡∏∞‡πÑ‡∏î‡πâ 1 ‡πÅ‡∏ñ‡∏ß)
        z_values = pivot_all.loc[symbol].values  # array ‡∏Ç‡∏≠‡∏á rolling_corr (‡∏ï‡∏≤‡∏°‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà pivot_all.columns)
        # ‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏≤‡∏Å z_values ‡πÄ‡∏õ‡πá‡∏ô 1D array (‡∏°‡∏µ len = ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà)
        # ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏Ç‡∏¢‡∏≤‡∏¢‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô 2D (pseudo-surface) ‡πÇ‡∏î‡∏¢‡∏ã‡πâ‡∏≥‡∏Ñ‡πà‡∏≤‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô 2 ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
        z_2d = np.array([z_values, z_values])  # shape = (2, nDates)
        
        # ‡πÅ‡∏Å‡∏ô X = pivot_all.columns (‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà), ‡πÅ‡∏Å‡∏ô Y = [symbol, symbol] (2 ‡πÅ‡∏ñ‡∏ß)
        x_2d = np.tile(pivot_all.columns.values, (2, 1))  # shape (2, nDates)
        y_2d = np.array([[symbol]*len(pivot_all.columns),
                         [symbol]*len(pivot_all.columns)])
        
        trace_symbol = go.Surface(
            x = x_2d,
            y = y_2d,
            z = z_2d,
            colorscale='rdbu',
            reversescale=True,
            connectgaps=True,
            opacity=0.95,
            showscale=False,  # ‡∏ã‡πà‡∏≠‡∏ô color bar ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö single symbol
            name=symbol
        )
        single_traces.append(trace_symbol)
    
    # ‡∏£‡∏ß‡∏° trace ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: trace_all ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å, ‡∏ï‡∏≤‡∏°‡∏î‡πâ‡∏ß‡∏¢ trace ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞ symbol
    all_traces = [trace_all] + single_traces
    
    # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏õ‡∏∏‡πà‡∏° Update Menu
    buttons = []
    
    # ‡∏õ‡∏∏‡πà‡∏° "All" -> ‡πÅ‡∏™‡∏î‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞ trace index=0
    visible_all = [True] + [False]*len(single_traces)
    buttons.append(dict(
        label="All",
        method="update",
        args=[
            {"visible": visible_all},
            {"title": "3D Surface Plot of Rolling Correlation (All)"}
        ]
    ))
    
    # ‡∏õ‡∏∏‡πà‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞ Symbol
    for i, symbol in enumerate(symbols):
        # ‡πÉ‡∏´‡πâ trace i+1 ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•, ‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠ False
        visibility = [False]*(1+len(single_traces))
        visibility[i+1] = True
        buttons.append(dict(
            label=symbol,
            method="update",
            args=[
                {"visible": visibility},
                {"title": f"3D Surface Plot of Rolling Correlation ({symbol})"}
            ]
        ))
    
    # Layout
    layout = go.Layout(
        title="3D Surface Plot of Rolling Correlation with Toggle Buttons",
        scene=dict(
            xaxis_title="Date",
            yaxis_title="Symbol",
            zaxis_title="Rolling Correlation",
            camera=dict(
                eye=dict(x=1.7, y=1.7, z=0.5)  # ‡∏õ‡∏£‡∏±‡∏ö‡∏°‡∏∏‡∏°‡∏°‡∏≠‡∏á‡∏ï‡∏≤‡∏°‡∏ä‡∏≠‡∏ö
            ),
        ),
        updatemenus=[{
            "buttons": buttons,
            "direction": "down",
            "showactive": True,
            "x": 0.1,
            "xanchor": "left",
            "y": 1.15,
            "yanchor": "top"
        }],
        margin=dict(t=80),
        annotations=[
            dict(
                text="Data Source: Your Data Source",
                x=0,
                y=-0.1,
                xref="paper",
                yref="paper",
                showarrow=False
            )
        ]
    )
    
    fig = go.Figure(data=all_traces, layout=layout)
    return fig

# -----------------------
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
# -----------------------
# df_result: DataFrame ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì rolling_corr ‡πÅ‡∏•‡πâ‡∏ß
# ‡πÇ‡∏î‡∏¢‡∏°‡∏µ index = date_only (‡∏´‡∏£‡∏∑‡∏≠ datetime) ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå: 'symbol', 'rolling_corr'
# ‡πÄ‡∏ä‡πà‡∏ô:
#   date_only   symbol        rolling_corr
# 2023-01-03  'NYSE:AAPL'      0.45
# 2023-01-04  'NYSE:AAPL'      0.43
# 2023-01-03  'NYSE:MSFT'      0.22
# ...          ...             ...
#
fig = surface_3d_with_toggle(df_result)
fig.show()
```
