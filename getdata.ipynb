```python
import os
import glob
import pandas as pd

# à¸à¸³à¸«à¸™à¸”à¹‚à¸Ÿà¸¥à¹€à¸”à¸­à¸£à¹Œà¸ªà¸³à¸«à¸£à¸±à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¸à¹‰à¸™à¹à¸¥à¸°à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Google Trend
stock_folder = 'ndx_date'   # à¹‚à¸Ÿà¸¥à¹€à¸”à¸­à¸£à¹Œà¹€à¸à¹‡à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¸à¹‰à¸™ Dow Jones 30 stocks
trend_folder = 'llmdate_copy'   # à¹‚à¸Ÿà¸¥à¹€à¸”à¸­à¸£à¹Œà¹€à¸à¹‡à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Google Trend
output_folder = 'output_folder' # à¹‚à¸Ÿà¸¥à¹€à¸”à¸­à¸£à¹Œà¸ªà¸³à¸«à¸£à¸±à¸šà¸šà¸±à¸™à¸—à¸¶à¸à¹„à¸Ÿà¸¥à¹Œà¸œà¸¥à¸¥à¸±à¸žà¸˜à¹Œ

os.makedirs(output_folder, exist_ok=True)

# ----- 1. à¸­à¹ˆà¸²à¸™à¹„à¸Ÿà¸¥à¹Œ CSV à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¸à¹‰à¸™ -----
stock_files = glob.glob(os.path.join(stock_folder, '*.csv'))
stock_dfs = []
for file in stock_files:
    # à¸­à¹ˆà¸²à¸™à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¸à¹‰à¸™ à¹‚à¸”à¸¢ parse à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'datetime'
    df = pd.read_csv(file, parse_dates=['datetime'])
    stock_dfs.append(df)
# à¸£à¸§à¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¸à¹‰à¸™à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
df_stock = pd.concat(stock_dfs, ignore_index=True)

# ----- 2. à¸­à¹ˆà¸²à¸™à¹„à¸Ÿà¸¥à¹Œ CSV à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Google Trend -----
trend_files = glob.glob(os.path.join(trend_folder, '*.csv'))
trend_dfs = []
for file in trend_files:
    # à¸­à¹ˆà¸²à¸™à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Google Trend à¹‚à¸”à¸¢ parse à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only'
    # à¸›à¸£à¸±à¸š dayfirst=True à¸«à¸²à¸à¸§à¸±à¸™à¸—à¸µà¹ˆà¸­à¸¢à¸¹à¹ˆà¹ƒà¸™à¸£à¸¹à¸›à¹à¸šà¸š dd/mm/yyyy (à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡: "3/1/2023" à¹ƒà¸«à¹‰à¹à¸›à¸¥à¹€à¸›à¹‡à¸™ 2023-01-03)
    df = pd.read_csv(file, parse_dates=['date_only'], dayfirst=True)
    trend_dfs.append(df)
# à¸£à¸§à¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Google Trend à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
df_trend = pd.concat(trend_dfs, ignore_index=True)

# ----- 3. à¸ªà¸£à¹‰à¸²à¸‡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only' à¸ªà¸³à¸«à¸£à¸±à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¸à¹‰à¸™ (à¸•à¸±à¸”à¹€à¸§à¸¥à¸²à¸ˆà¸²à¸ 'datetime') -----
df_stock['date_only'] = df_stock['datetime'].dt.date

# à¸ªà¸³à¸«à¸£à¸±à¸š Google Trend à¹€à¸£à¸²à¹à¸›à¸¥à¸‡à¹€à¸›à¹‡à¸™ date (à¸–à¹‰à¸²à¸¢à¸±à¸‡à¹„à¸¡à¹ˆà¹„à¸”à¹‰à¸—à¸³)
df_trend['date_only'] = df_trend['date_only'].dt.date

# ----- 4. Merge à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¸à¹‰à¸™à¸à¸±à¸š Google Trend à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only' -----
# à¹€à¸¥à¸·à¸­à¸à¹€à¸‰à¸žà¸²à¸°à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only' à¹à¸¥à¸° 'llm' à¸ˆà¸²à¸ df_trend
df_merged = pd.merge(df_stock, df_trend[['date_only', 'llm']], on='date_only', how='inner')

# à¹à¸›à¸¥à¸‡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only' à¹ƒà¸«à¹‰à¹€à¸›à¹‡à¸™ datetime (à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸Šà¹‰à¹ƒà¸™à¸à¸²à¸£ filter à¹à¸¥à¸° rolling)
df_merged['date_only'] = pd.to_datetime(df_merged['date_only'])
# à¸•à¸±à¹‰à¸‡ index à¹€à¸›à¹‡à¸™ 'date_only'
df_merged.set_index('date_only', inplace=True)

# ----- 5. à¸à¸£à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹ƒà¸«à¹‰à¸­à¸¢à¸¹à¹ˆà¹ƒà¸™à¸Šà¹ˆà¸§à¸‡à¸§à¸±à¸™à¸—à¸µà¹ˆà¸—à¸µà¹ˆà¸•à¹‰à¸­à¸‡à¸à¸²à¸£ -----
start_date = '2023-01-01'
end_date   = '2025-03-14'
df_merged = df_merged.loc[start_date:end_date]

# ----- 6. à¸„à¸³à¸™à¸§à¸“ Rolling Correlation (6 à¹€à¸”à¸·à¸­à¸™ ~ 180 à¸§à¸±à¸™)180D à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸«à¸¸à¹‰à¸™ -----
def compute_rolling_corr(group, window='180D'):
    group = group.sort_index()  # à¹€à¸£à¸µà¸¢à¸‡à¸¥à¸³à¸”à¸±à¸šà¸•à¸²à¸¡à¸§à¸±à¸™à¸—à¸µà¹ˆ
    # à¸„à¸³à¸™à¸§à¸“ rolling correlation à¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡ 'close' à¸à¸±à¸š 'llm'
    group['rolling_corr'] = group['close'].rolling(window=window).corr(group['llm'])
    return group

df_result = df_merged.groupby('symbol', group_keys=False).apply(compute_rolling_corr)

# à¹à¸ªà¸”à¸‡à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸œà¸¥à¸¥à¸±à¸žà¸˜à¹Œ
print("à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸µà¹ˆ merge à¹à¸¥à¸°à¸„à¸³à¸™à¸§à¸“ rolling correlation:")
print(df_result.head())

# ----- (à¸•à¸±à¸§à¹€à¸¥à¸·à¸­à¸) à¸šà¸±à¸™à¸—à¸¶à¸à¸œà¸¥à¸¥à¸±à¸žà¸˜à¹Œà¸¥à¸‡à¹„à¸Ÿà¸¥à¹Œ CSV -----
output_filepath = os.path.join(output_folder, 'merged_rolling_corr_adx.csv')
df_result.to_csv(output_filepath)
print(f"\nSaved merged and rolling correlation data to {output_filepath}")
```

```python
import plotly.graph_objects as go
import pandas as pd

def surface_3d_rolling_corr(df):
    """
    à¸ªà¸£à¹‰à¸²à¸‡ 3D surface plot à¸ªà¸³à¸«à¸£à¸±à¸š rolling correlation
    à¹‚à¸”à¸¢ df à¸„à¸§à¸£à¸¡à¸µà¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only', 'symbol' à¹à¸¥à¸° 'rolling_corr'
    """
    # Pivot DataFrame à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¹„à¸”à¹‰ grid à¸ªà¸³à¸«à¸£à¸±à¸š plot:
    # index (Y-axis): date_only, columns (X-axis): symbol, values (Z-axis): rolling_corr
    pivot_df = df.reset_index().pivot(index='date_only', columns='symbol', values='rolling_corr')
    
    # à¸ªà¸£à¹‰à¸²à¸‡ figure 3D surface plot
    fig = go.Figure(data=[go.Surface(
        x=pivot_df.columns,       # à¸£à¸²à¸¢à¸Šà¸·à¹ˆà¸­à¸«à¸¸à¹‰à¸™
        y=pivot_df.index,         # à¸§à¸±à¸™à¸—à¸µà¹ˆ
        z=pivot_df.values,        # à¸„à¹ˆà¸² rolling correlation
        opacity=0.95,
        connectgaps=True,
        colorscale='rdbu',
        showscale=True,
        reversescale=True,
    )])
    
    # à¸›à¸£à¸±à¸š layout à¸‚à¸­à¸‡à¸à¸£à¸²à¸Ÿ
    fig.update_layout(
        title='3D Surface Plot of Rolling Correlation',
        title_font=dict(size=20),
        autosize=True,
        height=500,
        hovermode='closest',
        scene={
            "aspectratio": {"x": 1, "y": 2.2, "z": 1},
            'camera': {'eye': {'x': 2, 'y': 0.4, 'z': 0.8}},
            'xaxis_title': 'Symbol',
            'yaxis_title': 'Date',
            'zaxis_title': 'Rolling Correlation'
        },
        margin=dict(t=40),
        annotations=[
            dict(
                text="Data Source: Your Data Source",
                x=0,
                y=-0.15,
                xref="paper",
                yref="paper",
                showarrow=False
            )
        ]
    )
    return fig

# à¸ªà¸¡à¸¡à¸¸à¸•à¸´à¸§à¹ˆà¸² df_result à¸„à¸·à¸­ DataFrame à¸—à¸µà¹ˆà¸¡à¸µ rolling_corr à¸žà¸£à¹‰à¸­à¸¡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only', 'symbol'
# à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸à¸²à¸£à¹€à¸£à¸µà¸¢à¸à¹ƒà¸Šà¹‰à¸‡à¸²à¸™:
fig = surface_3d_rolling_corr(df_result)
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

def create_3d_surface_with_toggle(pivot_df):
    """
    à¸ªà¸£à¹‰à¸²à¸‡à¸à¸£à¸²à¸Ÿ 3D à¸—à¸µà¹ˆà¸¡à¸µ update menu à¹ƒà¸«à¹‰à¹€à¸¥à¸·à¸­à¸à¸”à¸¹ rolling correlation à¸‚à¸­à¸‡à¸«à¸¸à¹‰à¸™
    pivot_df à¸„à¸§à¸£à¹€à¸›à¹‡à¸™ DataFrame à¸—à¸µà¹ˆ pivot à¹‚à¸”à¸¢à¸¡à¸µ index à¹€à¸›à¹‡à¸™à¸§à¸±à¸™à¸—à¸µà¹ˆ
    à¹à¸¥à¸° columns à¹€à¸›à¹‡à¸™à¸Šà¸·à¹ˆà¸­à¸«à¸¸à¹‰à¸™ (symbol) à¹‚à¸”à¸¢à¸„à¹ˆà¸²à¸—à¸µà¹ˆà¸­à¸¢à¸¹à¹ˆà¹ƒà¸™ pivot à¸„à¸·à¸­ rolling_corr
    """
    # à¸”à¸¶à¸‡à¸Šà¸·à¹ˆà¸­à¸«à¸¸à¹‰à¸™à¸ˆà¸²à¸ pivot table
    symbols = list(pivot_df.columns)
    # à¸à¸³à¸«à¸™à¸”à¹à¸à¸™ X à¹à¸šà¸šà¸•à¸±à¸§à¹€à¸¥à¸‚à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸«à¸¸à¹‰à¸™
    x_numeric = np.arange(len(symbols))
    
    # --- Trace à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸ªà¸”à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ "All" ---
    all_trace = go.Surface(
        x = x_numeric,
        y = pivot_df.index,
        z = pivot_df.values,
        opacity=0.95,
        colorscale='rdbu',
        reversescale=True,
        showscale=True,
        name="All"
    )
    
    traces = [all_trace]
    
    # --- Trace à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸«à¸¸à¹‰à¸™ (Scatter3d à¹à¸šà¸šà¹€à¸ªà¹‰à¸™) ---
    for i, symbol in enumerate(symbols):
        scatter_trace = go.Scatter3d(
            x = [x_numeric[i]] * len(pivot_df.index),
            y = pivot_df.index,
            z = pivot_df[symbol].values,
            mode='lines',
            name=symbol
        )
        traces.append(scatter_trace)
    
    # à¸ªà¸£à¹‰à¸²à¸‡ Figure à¸žà¸£à¹‰à¸­à¸¡à¸à¸±à¸š trace à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
    fig = go.Figure(data=traces)
    
    # --- à¸ªà¸£à¹‰à¸²à¸‡ update menu buttons ---
    buttons = []
    # à¸›à¸¸à¹ˆà¸¡ "All" à¹ƒà¸«à¹‰à¹à¸ªà¸”à¸‡à¹€à¸‰à¸žà¸²à¸° trace à¹à¸šà¸š Surface (trace à¸—à¸µà¹ˆ 0)
    visibility_all = [True] + [False] * len(symbols)
    buttons.append(dict(
        label="All",
        method="update",
        args=[{"visible": visibility_all},
              {"title": "Rolling Correlation - All Stocks"}]
    ))
    
    # à¸›à¸¸à¹ˆà¸¡à¸ªà¸³à¸«à¸£à¸±à¸šà¸«à¸¸à¹‰à¸™à¹à¸•à¹ˆà¸¥à¸°à¸•à¸±à¸§ (trace à¸—à¸µà¹ˆ 1 à¸–à¸¶à¸‡ n)
    for i, symbol in enumerate(symbols):
        # à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸›à¸¸à¹ˆà¸¡ à¹ƒà¸«à¹‰à¸‹à¹ˆà¸­à¸™ trace à¹à¸šà¸š Surface (index 0) à¹à¸¥à¸°à¹à¸ªà¸”à¸‡à¹€à¸‰à¸žà¸²à¸° trace à¸‚à¸­à¸‡à¸«à¸¸à¹‰à¸™à¸™à¸±à¹‰à¸™
        visibility = [False]  # à¸‹à¹ˆà¸­à¸™ trace "All"
        # à¸ªà¸³à¸«à¸£à¸±à¸š scatter traces: à¹ƒà¸«à¹‰ trace à¸—à¸µà¹ˆ i+1 (à¹€à¸žà¸£à¸²à¸°à¹€à¸£à¸´à¹ˆà¸¡à¸ˆà¸²à¸ index 1) à¹€à¸›à¹‡à¸™ True
        for j in range(len(symbols)):
            visibility.append(j == i)
        buttons.append(dict(
            label=symbol,
            method="update",
            args=[{"visible": visibility},
                  {"title": f"Rolling Correlation - {symbol}"}]
        ))
    
    fig.update_layout(
        updatemenus=[{
            "buttons": buttons,
            "direction": "down",
            "showactive": True,
            "x": 0.1,
            "xanchor": "left",
            "y": 1.1,
            "yanchor": "top"
        }],
        title="Rolling Correlation 3D Plot",
        scene=dict(
            xaxis=dict(
                title="Stock Symbol",
                tickmode="array",
                tickvals=x_numeric,
                ticktext=symbols
            ),
            yaxis_title="Date",
            zaxis_title="Rolling Correlation"
        )
    )
    
    return fig

# à¸ªà¸¡à¸¡à¸¸à¸•à¸´à¸§à¹ˆà¸² pivot_df à¸„à¸·à¸­ DataFrame à¸—à¸µà¹ˆà¹„à¸”à¹‰à¸ˆà¸²à¸à¸à¸²à¸£ pivot à¸‚à¹‰à¸­à¸¡à¸¹à¸¥
# à¹€à¸Šà¹ˆà¸™ à¸ˆà¸²à¸ DataFrame df_result à¸—à¸µà¹ˆà¸¡à¸µà¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'date_only', 'symbol' à¹à¸¥à¸° 'rolling_corr'
# à¹€à¸£à¸²à¸ªà¸²à¸¡à¸²à¸£à¸–à¸ªà¸£à¹‰à¸²à¸‡ pivot_df à¹„à¸”à¹‰à¸”à¹‰à¸§à¸¢à¸„à¸³à¸ªà¸±à¹ˆà¸‡:
#
#   pivot_df = df_result.reset_index().pivot(index='date_only', columns='symbol', values='rolling_corr')
#
# à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸™à¸µà¹‰à¹ƒà¸Šà¹‰ pivot_df à¸—à¸µà¹ˆà¸¡à¸µ index à¹€à¸›à¹‡à¸™à¸§à¸±à¸™à¸—à¸µà¹ˆ (DatetimeIndex) à¹à¸¥à¸° columns à¹€à¸›à¹‡à¸™à¸Šà¸·à¹ˆà¸­à¸«à¸¸à¹‰à¸™

# à¹€à¸£à¸µà¸¢à¸à¹ƒà¸Šà¹‰à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¹€à¸žà¸·à¹ˆà¸­à¸ªà¸£à¹‰à¸²à¸‡à¸à¸£à¸²à¸Ÿ 3D à¸žà¸£à¹‰à¸­à¸¡à¸›à¸¸à¹ˆà¸¡à¹€à¸›à¸´à¸”/à¸›à¸´à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¹€à¸¥à¸·à¸­à¸à¸«à¸¸à¹‰à¸™
fig = create_3d_surface_with_toggle(df_result)
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

def plot_3d_lines_with_toggle(df_result):
    """
    à¸ªà¸£à¹‰à¸²à¸‡ 3D line plot à¸ªà¸³à¸«à¸£à¸±à¸š rolling correlation à¹‚à¸”à¸¢à¸¡à¸µà¸›à¸¸à¹ˆà¸¡à¸ªà¸¥à¸±à¸šà¸à¸²à¸£à¹à¸ªà¸”à¸‡à¸œà¸¥à¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸°à¸«à¸¸à¹‰à¸™
    df_result à¸„à¸§à¸£à¸¡à¸µ index à¹€à¸›à¹‡à¸™à¸§à¸±à¸™à¸—à¸µà¹ˆ à¹à¸¥à¸°à¸¡à¸µà¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'symbol' à¹à¸¥à¸° 'rolling_corr'
    """
    # à¸”à¸¶à¸‡à¸£à¸²à¸¢à¸Šà¸·à¹ˆà¸­à¸«à¸¸à¹‰à¸™à¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¸‹à¹‰à¸³ à¹à¸¥à¹‰à¸§à¸ˆà¸±à¸”à¹€à¸£à¸µà¸¢à¸‡à¹€à¸žà¸·à¹ˆà¸­à¸„à¸§à¸²à¸¡à¹à¸™à¹ˆà¹ƒà¸ˆ
    symbols = sorted(df_result['symbol'].unique())
    # à¸à¸³à¸«à¸™à¸” mapping à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸à¸™ y: à¹à¸•à¹ˆà¸¥à¸°à¸«à¸¸à¹‰à¸™à¸ˆà¸°à¹„à¸”à¹‰à¸„à¹ˆà¸² y à¹€à¸‰à¸žà¸²à¸° (0, 1, 2, ...)
    symbol_to_y = {symbol: i for i, symbol in enumerate(symbols)}
    
    # à¸ªà¸£à¹‰à¸²à¸‡ trace à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸«à¸¸à¹‰à¸™ (Scatter3d à¹à¸šà¸š line)
    traces = []
    for symbol in symbols:
        # à¸à¸£à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸ªà¸³à¸«à¸£à¸±à¸šà¸«à¸¸à¹‰à¸™à¸™à¸±à¹‰à¸™
        df_sym = df_result[df_result['symbol'] == symbol].sort_index()
        # à¸à¸³à¸«à¸™à¸”à¹à¸à¸™ x à¹€à¸›à¹‡à¸™à¸§à¸±à¸™à¸—à¸µà¹ˆ, à¹à¸à¸™ y à¹€à¸›à¹‡à¸™à¸„à¹ˆà¸²à¸—à¸µà¹ˆà¸à¸³à¸«à¸™à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¸«à¸¸à¹‰à¸™à¸™à¸±à¹‰à¸™, à¹à¸à¸™ z à¹€à¸›à¹‡à¸™ rolling_corr
        x = df_sym.index  # index à¹€à¸›à¹‡à¸™à¸§à¸±à¸™à¸—à¸µà¹ˆ
        y = [symbol_to_y[symbol]] * len(df_sym)
        z = df_sym['rolling_corr']
        
        trace = go.Scatter3d(
            x=x,
            y=y,
            z=z,
            mode='lines+markers',
            name=symbol
        )
        traces.append(trace)
    
    # à¸ªà¸£à¹‰à¸²à¸‡à¸›à¸¸à¹ˆà¸¡à¸ªà¸³à¸«à¸£à¸±à¸šà¸ªà¸¥à¸±à¸šà¸à¸²à¸£à¹à¸ªà¸”à¸‡à¸œà¸¥ (update menu)
    buttons = []
    
    # à¸›à¸¸à¹ˆà¸¡ "All" à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸ªà¸”à¸‡à¸—à¸¸à¸à¸«à¸¸à¹‰à¸™
    all_visible = [True] * len(traces)
    buttons.append(dict(
        label="All",
        method="update",
        args=[{"visible": all_visible},
              {"title": "Rolling Correlation: All Stocks"}]
    ))
    
    # à¸›à¸¸à¹ˆà¸¡à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸«à¸¸à¹‰à¸™à¹à¸¢à¸à¸à¸±à¸™
    for i, symbol in enumerate(symbols):
        visibility = [False] * len(traces)
        visibility[i] = True
        buttons.append(dict(
            label=symbol,
            method="update",
            args=[{"visible": visibility},
                  {"title": f"Rolling Correlation: {symbol}"}]
        ))
    
    layout = go.Layout(
        title="3D Rolling Correlation with Toggle Buttons",
        scene=dict(
            xaxis_title="Date",
            yaxis_title="Stock",
            zaxis_title="Rolling Correlation",
            # à¸à¸³à¸«à¸™à¸” tick labels à¹ƒà¸«à¹‰à¹à¸à¸™ y à¸•à¸²à¸¡ mapping
            yaxis=dict(
                tickmode='array',
                tickvals=list(symbol_to_y.values()),
                ticktext=list(symbol_to_y.keys())
            )
        ),
        updatemenus=[{
            "buttons": buttons,
            "direction": "down",
            "showactive": True,
            "x": 0.1,
            "xanchor": "left",
            "y": 1.1,
            "yanchor": "top"
        }]
    )
    
    fig = go.Figure(data=traces, layout=layout)
    return fig

# à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸à¸²à¸£à¹€à¸£à¸µà¸¢à¸à¹ƒà¸Šà¹‰à¸‡à¸²à¸™
# à¸ªà¸¡à¸¡à¸¸à¸•à¸´à¸§à¹ˆà¸² df_result à¸„à¸·à¸­ DataFrame à¸—à¸µà¹ˆà¹„à¸”à¹‰ merge à¹à¸¥à¸°à¸„à¸³à¸™à¸§à¸“ rolling correlation à¹„à¸§à¹‰à¹à¸¥à¹‰à¸§
# à¹‚à¸”à¸¢ index à¹€à¸›à¹‡à¸™ datetime (à¸§à¸±à¸™à¸—à¸µà¹ˆ) à¹à¸¥à¸°à¸¡à¸µà¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'symbol' à¹à¸¥à¸° 'rolling_corr'

# à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸à¸²à¸£à¹€à¸£à¸µà¸¢à¸à¹ƒà¸Šà¹‰à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™
fig = plot_3d_lines_with_toggle(df_result)
fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

def surface_3d_with_toggle(df_result):
    """
    à¸ªà¸£à¹‰à¸²à¸‡ 3D Surface Plot à¸‚à¸­à¸‡ Rolling Correlation
    à¸žà¸£à¹‰à¸­à¸¡à¸›à¸¸à¹ˆà¸¡ (Toggle) à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸ªà¸”à¸‡ 'All' à¸«à¸£à¸·à¸­ 'Single Symbol'
    
    - df_result à¸„à¸§à¸£à¸¡à¸µ:
        index: à¹€à¸›à¹‡à¸™ date (date_only à¸«à¸£à¸·à¸­ datetime)
        columns: 'symbol', 'rolling_corr'
    """
    # à¸£à¸µà¹€à¸‹à¹‡à¸• index à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¹à¸™à¹ˆà¹ƒà¸ˆà¸§à¹ˆà¸²à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ date à¸¡à¸µà¸Šà¸·à¹ˆà¸­à¸Šà¸±à¸”à¹€à¸ˆà¸™
    df_temp = df_result.reset_index().copy()
    # à¸•à¸±à¹‰à¸‡à¸Šà¸·à¹ˆà¸­à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ index à¸‚à¸­à¸‡à¸§à¸±à¸™à¸—à¸µà¹ˆà¹ƒà¸«à¹‰à¸Šà¸±à¸”à¹€à¸ˆà¸™ (à¸–à¹‰à¸²à¹ƒà¸™ df_result à¹€à¸›à¹‡à¸™ date_only à¸­à¸¢à¸¹à¹ˆà¹à¸¥à¹‰à¸§ à¹ƒà¸«à¹‰à¸›à¸£à¸±à¸šà¸•à¸²à¸¡à¸ˆà¸£à¸´à¸‡)
    df_temp.rename(columns={'index': 'date_only'}, inplace=True)
    
    # Pivot à¹€à¸žà¸·à¹ˆà¸­à¸—à¸³ Surface "All":
    #   index  = symbol
    #   columns= date
    #   values = rolling_corr
    pivot_all = df_temp.pivot(index='symbol', columns='date_only', values='rolling_corr')
    pivot_all = pivot_all.sort_index(axis=0)  # à¹€à¸£à¸µà¸¢à¸‡ symbol
    pivot_all = pivot_all.sort_index(axis=1)  # à¹€à¸£à¸µà¸¢à¸‡ date
    
    # à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸§à¹ˆà¸²à¸žà¸­à¸¡à¸µà¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸£à¸·à¸­à¹„à¸¡à¹ˆ
    if pivot_all.empty:
        raise ValueError("Pivot DataFrame à¸§à¹ˆà¸²à¸‡à¹€à¸›à¸¥à¹ˆà¸² à¸à¸£à¸¸à¸“à¸²à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥ df_result")
    
    # à¸ªà¸£à¹‰à¸²à¸‡ trace à¸ªà¸³à¸«à¸£à¸±à¸š "All" (Surface à¹€à¸”à¸µà¸¢à¸§à¸—à¸µà¹ˆà¸¡à¸µ Symbol à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”)
    trace_all = go.Surface(
        x = pivot_all.columns,        # X = Date
        y = pivot_all.index,          # Y = Symbol
        z = pivot_all.values,         # Z = Rolling Correlation
        colorscale='rdbu',
        reversescale=True,
        connectgaps=True,
        opacity=0.95,
        showscale=True,    # à¹à¸ªà¸”à¸‡ color bar à¸ªà¸³à¸«à¸£à¸±à¸š All
        name='All'
    )
    
    # à¹€à¸•à¸£à¸µà¸¢à¸¡à¸ªà¸£à¹‰à¸²à¸‡ surface à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸° Symbol
    # à¹‚à¸”à¸¢à¸ªà¸³à¸«à¸£à¸±à¸š symbol à¹ƒà¸”à¹† à¹€à¸¡à¸·à¹ˆà¸­ fix symbol à¹à¸¥à¹‰à¸§ à¸ˆà¸°à¹€à¸«à¸¥à¸·à¸­à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ date vs rolling_corr à¹€à¸›à¹‡à¸™ 1 à¹à¸–à¸§
    # à¸•à¹‰à¸­à¸‡à¸—à¸³ pseudo-surface 2 à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ
    single_traces = []
    symbols = pivot_all.index.tolist()  # à¸£à¸²à¸¢à¸Šà¸·à¹ˆà¸­à¸«à¸¸à¹‰à¸™à¸ˆà¸²à¸ pivot
    
    for symbol in symbols:
        # à¸”à¸¶à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ rolling_corr à¸‚à¸­à¸‡ symbol à¸™à¸±à¹‰à¸™ (à¸ˆà¸°à¹„à¸”à¹‰ 1 à¹à¸–à¸§)
        z_values = pivot_all.loc[symbol].values  # array à¸‚à¸­à¸‡ rolling_corr (à¸•à¸²à¸¡à¸§à¸±à¸™à¸—à¸µà¹ˆ pivot_all.columns)
        # à¹€à¸™à¸·à¹ˆà¸­à¸‡à¸ˆà¸²à¸ z_values à¹€à¸›à¹‡à¸™ 1D array (à¸¡à¸µ len = à¸ˆà¸³à¸™à¸§à¸™à¸§à¸±à¸™à¸—à¸µà¹ˆ)
        # à¹€à¸£à¸²à¸ˆà¸°à¸‚à¸¢à¸²à¸¢à¹ƒà¸«à¹‰à¹€à¸›à¹‡à¸™ 2D (pseudo-surface) à¹‚à¸”à¸¢à¸‹à¹‰à¸³à¸„à¹ˆà¸²à¸™à¸µà¹‰à¹€à¸›à¹‡à¸™ 2 à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ
        z_2d = np.array([z_values, z_values])  # shape = (2, nDates)
        
        # à¹à¸à¸™ X = pivot_all.columns (à¸§à¸±à¸™à¸—à¸µà¹ˆ), à¹à¸à¸™ Y = [symbol, symbol] (2 à¹à¸–à¸§)
        x_2d = np.tile(pivot_all.columns.values, (2, 1))  # shape (2, nDates)
        y_2d = np.array([[symbol]*len(pivot_all.columns),
                         [symbol]*len(pivot_all.columns)])
        
        trace_symbol = go.Surface(
            x = x_2d,
            y = y_2d,
            z = z_2d,
            colorscale='rdbu',
            reversescale=True,
            connectgaps=True,
            opacity=0.95,
            showscale=False,  # à¸‹à¹ˆà¸­à¸™ color bar à¸ªà¸³à¸«à¸£à¸±à¸š single symbol
            name=symbol
        )
        single_traces.append(trace_symbol)
    
    # à¸£à¸§à¸¡ trace à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”: trace_all à¹€à¸›à¹‡à¸™à¸•à¸±à¸§à¹à¸£à¸, à¸•à¸²à¸¡à¸”à¹‰à¸§à¸¢ trace à¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸° symbol
    all_traces = [trace_all] + single_traces
    
    # à¸ªà¸£à¹‰à¸²à¸‡à¸›à¸¸à¹ˆà¸¡ Update Menu
    buttons = []
    
    # à¸›à¸¸à¹ˆà¸¡ "All" -> à¹à¸ªà¸”à¸‡à¹€à¸‰à¸žà¸²à¸° trace index=0
    visible_all = [True] + [False]*len(single_traces)
    buttons.append(dict(
        label="All",
        method="update",
        args=[
            {"visible": visible_all},
            {"title": "3D Surface Plot of Rolling Correlation (All)"}
        ]
    ))
    
    # à¸›à¸¸à¹ˆà¸¡à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸° Symbol
    for i, symbol in enumerate(symbols):
        # à¹ƒà¸«à¹‰ trace i+1 à¹à¸ªà¸”à¸‡à¸œà¸¥, à¸—à¸µà¹ˆà¹€à¸«à¸¥à¸·à¸­ False
        visibility = [False]*(1+len(single_traces))
        visibility[i+1] = True
        buttons.append(dict(
            label=symbol,
            method="update",
            args=[
                {"visible": visibility},
                {"title": f"3D Surface Plot of Rolling Correlation ({symbol})"}
            ]
        ))
    
    # Layout
    layout = go.Layout(
        title="3D Surface Plot of Rolling Correlation with Toggle Buttons",
        scene=dict(
            xaxis_title="Date",
            yaxis_title="Symbol",
            zaxis_title="Rolling Correlation",
            camera=dict(
                eye=dict(x=1.7, y=1.7, z=0.5)  # à¸›à¸£à¸±à¸šà¸¡à¸¸à¸¡à¸¡à¸­à¸‡à¸•à¸²à¸¡à¸Šà¸­à¸š
            ),
        ),
        updatemenus=[{
            "buttons": buttons,
            "direction": "down",
            "showactive": True,
            "x": 0.1,
            "xanchor": "left",
            "y": 1.15,
            "yanchor": "top"
        }],
        margin=dict(t=80),
        annotations=[
            dict(
                text="Data Source: Your Data Source",
                x=0,
                y=-0.1,
                xref="paper",
                yref="paper",
                showarrow=False
            )
        ]
    )
    
    fig = go.Figure(data=all_traces, layout=layout)
    return fig

# -----------------------
# à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸à¸²à¸£à¹ƒà¸Šà¹‰à¸‡à¸²à¸™
# -----------------------
# df_result: DataFrame à¸—à¸µà¹ˆà¸œà¹ˆà¸²à¸™à¸à¸²à¸£à¸„à¸³à¸™à¸§à¸“ rolling_corr à¹à¸¥à¹‰à¸§
# à¹‚à¸”à¸¢à¸¡à¸µ index = date_only (à¸«à¸£à¸·à¸­ datetime) à¹à¸¥à¸°à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ: 'symbol', 'rolling_corr'
# à¹€à¸Šà¹ˆà¸™:
#   date_only   symbol        rolling_corr
# 2023-01-03  'NYSE:AAPL'      0.45
# 2023-01-04  'NYSE:AAPL'      0.43
# 2023-01-03  'NYSE:MSFT'      0.22
# ...          ...             ...
#
fig = surface_3d_with_toggle(df_result)
fig.show()
```
